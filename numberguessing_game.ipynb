```python
import random  # To generate a random number
```

```python
def number_guessing_game():
    # Computer selects a random number between 1 and 100
    number_to_guess = random.randint(1, 100)

    attempts = 0  # Count the number of guesses
    guessed = False  # To check if the user has guessed correctly

    print("Welcome to the Number Guessing Game!")
    print("I have selected a number between 1 and 100. Can you guess it?")

    # Keep asking the user until they guess correctly
    while not guessed:
        try:
            user_guess = int(input("Enter your guess: "))  # Take input from the user
            attempts += 1  # Increase attempt count

            if user_guess < number_to_guess:
                print("Too low! Try a higher number.")  # Hint: guess is too low
            elif user_guess > number_to_guess:
                print("Too high! Try a lower number.")  # Hint: guess is too high
            else:
                guessed = True  # User guessed the number
                print(f"Congratulations! You guessed the number {number_to_guess} in {attempts} attempts.")
        except ValueError:
            print("Please enter a valid number.")  # Handle non-number input

number_guessing_game()  # Call the game function to play once

while True:
    number_guessing_game()  # Play the game
    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again != "yes":
        print("Thanks for playing! Goodbye!")
        break
```
